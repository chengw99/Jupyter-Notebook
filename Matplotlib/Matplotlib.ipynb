## **2. 画图 **

### 2.1 概念

* 一个Figure对象可以包含多个子图（Axes）

### 2.2 绘图的基本步骤

```python
x = np.linspace(0,10,1000)
y = np.sin(x)
z = np.cos(x**2)

# 创建一个Figure对象，并且它成为当前Figure对象
plt.figure(figsize=(8,4)) # figsize参数指定Figure对象的宽度和高度，单位为英尺
# 此外，还可用dpi参数指定Figure对象的分辨率，即每英寸所表示的像素数。默认值为80

# 调用plot() 在当前的Figure对象中绘图
# 实际上plot()是在Axes(子图)对象上绘图，如果当前的Figure对象中没有Axes对象，将会为之创建一个几乎充满整个图表的Axes对象
plt.plot(x,y,label="$sin(x)$",color='red',linewidth=2) # label 如果标签字的前后有字符'$',会使用内嵌的LaTex显示数学公式
plt.plot(x,z,"b--",label='$cons(x^2)$') 

plt.xlabel('Time(s)')
plt.ylabel('Volt')
plt.title('Pyplot First example')
plt.ylim(-1.2,1.2)
plt.legend()
#plt.show() 在Notebook中可以省略这个步骤
```

### 2.3 绘制多个子图 

#### 2.31 子图均匀分布

可以使用subplot()快速绘制包含多个子图的图表，调用形式如下：
***
**subplot(numRows,numCols,plotnum）**
***

* <font color=Crismon>图表的整个绘图区域被等分为numRows行和numCols列, plotNum参数指定所创建Axes对象的区域</font>, 然后按照<font color=DarkViolet>**从左到右、从上到下**</font>的顺序对每个区域进行编号，左上区域的编号为1.
* 如果numRows、numCols和plotNum三个参数都小于10，则可以把它们缩写成一个整数，例如subplot(323)和 subplot(3,2,3)的含义相同.
* 如果新创建的子图和之前创建的子图区域有重叠的部分，之前的子图将被删除.

```python
plt.subplot(221) # 分成2x2，占用第一个
plt.subplot(222) # 分成2x2，占用第二个
plt.subplot(212) # 分成2x1，占用第二个，即第二行
```

演示在不同图表的不同子图中绘制曲线

```python
plt.figure(1) # 创建图表1
plt.figure(2) # 创建图表2
ax1 = plt.subplot(121) # 在图表2中创建子图1
ax2 = plt.subplot(122) # 在图表2中创建子图2

x = np.linspace(0,3,100)
for i in range(5):
    plt.figure(1)
    plt.plot(x,np.exp(i*x/3))
    plt.sca(ax1) # 选择图表2的子图1
    plt.plot(x,np.sin(i*x))
    plt.sca(ax2) # 选择图表2的子图2
    plt.plot(x,np.cos(i*x))

```

#### 2.32 子图不均匀分布 

***
subplot2grid(shape,loc,rowspan=1,colspan=1,**kwargs)
***

* shape为表示表格形状的元组：(行数，列数)
* loc为子图左上角所在的坐标：(行，列)
* rowspan和colspan分別为子图所占据的行数和列数

```python
fig = plt.figure(figsize=(12,9))
ax1 = plt.subplot2grid((3,3),(0,0),colspan=2)
ax2 = plt.subplot2grid((3,3),(0,2),rowspan=2)
ax3 = plt.subplot2grid((3,3),(1,0),rowspan=2)
ax4 = plt.subplot2grid((3,3),(2,1),colspan=2)
ax5 = plt.subplot2grid((3,3),(1,1))
```
