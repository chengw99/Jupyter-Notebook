## 2. 画图 

### 2.1 绘图的基本步骤

```python
x = np.linspace(0,10,1000)
y = np.sin(x)
z = np.cos(x**2)

# 创建一个Figure对象，并且它成为当前Figure对象
plt.figure(figsize=(8,4)) # figsize参数指定Figure对象的宽度和高度，单位为英尺
# 此外，还可用dpi参数指定Figure对象的分辨率，即每英寸所表示的像素数。默认值为80

# 调用plot() 在当前的Figure对象中绘图
# 实际上plot()是在Axes(子图)对象上绘图，如果当前的Figure对象中没有Axes对象，将会为之创建一个几乎充满整个图表的Axes对象
plt.plot(x,y,label="$sin(x)$",color='red',linewidth=2) # label 如果标签字的前后有字符'$',会使用内嵌的LaTex显示数学公式
plt.plot(x,z,"b--",label='$cons(x^2)$') 

plt.xlabel('Time(s)')
plt.ylabel('Volt')
plt.title('Pyplot First example')
plt.ylim(-1.2,1.2)
plt.legend()
#plt.show() 在Notebook中可以省略这个步骤
```
